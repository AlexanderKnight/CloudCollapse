```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.animation as animation
import copy
import time
import datetime
from mpl_toolkits.mplot3d import Axes3D



class Clump:
    
    
    def __init__(self, Space=1000, MPassed=1000, V=0, color=0):
        self.x = np.random.rand()*Space-np.random.rand()*Space
        self.y = np.random.rand()*Space-np.random.rand()*Space
        self.z = np.random.rand()*Space-np.random.rand()*Space
        self.vx = np.random.rand()*V-np.random.rand()*V
        self.vy = np.random.rand()*V-np.random.rand()*V
        self.vz = np.random.rand()*V-np.random.rand()*V
        self.mass = np.random.rand()*MPassed
        self.radius = (self.mass/((4/3)*np.pi*1392))**(1/3)
        self.color = plt.cm.bone(color)
        
    def copy(self):
        newClump = copy.deepcopy(self)
        return newClump
        
        
    def distance(self, Clump2):
        return np.sqrt((Clump2.x-self.x)**2+(Clump2.y-self.y)**2+(Clump2.z-self.z)**2)
    
    def acc(self,Clump2,time):
        self.vx += (6.67408*10**(-11))*Clump2.mass/(self.distance(Clump2))**(3/2)*(Clump2.x-self.x)*time
        self.vy += (6.67408*10**(-11))*Clump2.mass/(self.distance(Clump2))**(3/2)*(Clump2.y-self.y)*time
        self.vz += (6.67408*10**(-11))*Clump2.mass/(self.distance(Clump2))**(3/2)*(Clump2.z-self.z)*time
        
    def move(self,time):
        self.x += self.vx*time
        self.y += self.vy*time
        self.z += self.vz*time
    
    def merge(self,Clump2):
        newClump = Clump()
        newClump.x = (Clump2.x+self.x)/2
        newClump.y = (Clump2.y+self.y)/2
        newClump.z = (Clump2.z+self.z)/2
        newClump.vx = self.vx+Clump2.vx
        newClump.vy = self.vy+Clump2.vy
        newClump.vz = self.vz+Clump2.vz
        newClump.mass = self.mass+Clump2.mass
        return newClump
        
    def drawClump(self, fig):
        clump = plt.Circle((self.x,self.y),self.radius)
        fig.gca().add_artist(clump)
        
        
    def x(self):
        return self.x
    def y(self):
        return self.y
    def z(self):
        return self.z
    def vx(self):
        return vx
    def vy(self):
        return vy
    def vz(self):
        return vz

    
    
class Cloud:
    
    
    
    def __init__(self, N=100, Mass = 1000, Space = 1000, V=0, color = 0, new=True, c=[]):
        if new ==True:
            self.clumps=[]
            for i in range(N):
                self.clumps.append(Clump(Space, Mass, V, color))
            self.N = N
        else:
            self.clumps = c
            
            self.N = N
        
    def __getitem__(self,index):
        return self.clumps[index]
    
    def copy(self):
        newCloud = copy.deepcopy(self)
        return newCloud
        
    
    def timeStep(self, time):
        k = 0
        l = k+1
        while k < self.N-1:
            while l <self.N:
                                             
                if self.clumps[k].distance(self.clumps[l]) <= (self.clumps[k].radius + self.clumps[l].radius) :
                    newClump = self.clumps[k].merge(self.clumps[l])
                    self.remove(self.clumps[l])
                    self.remove(self.clumps[k])
                    self.clumps.append(newClump)
                    
                    self.N = self.N+1
                l+=1
            k+=1
            l=k+1
            
        for i in range(self.N-1):
            for j in range(i+1,self.N):
                self.clumps[i].acc(self.clumps[j], time)
                self.clumps[j].acc(self.clumps[i], time)
                
        for i in range(self.N):
            self.clumps[i].move(time)
            
        return self.clumps
    
    
    def status(self):
        X = []
        Y = []
        Z = []
        Mass = []
        Radius = []
        for i in range(self.N):
            tempClump = self.clumps[i]
            X.append(tempClump.x)
            Y.append(tempClump.y)
            Z.append(tempClump.z)
            Mass.append(tempClump.mass)
            Radius.append(tempClump.radius)
        return X, Y, Z, Mass, Radius
    
    def remove(self, item):
        tempClumps = []
        for i in range(self.N):
            if i != item:
                tempClumps.append(self.clumps[i])
        self.clumps = tempClumps
        self.N = self.N -1
    
    def cm(self):
        M=0
        mx = 0
        my = 0
        mz = 0
        for i in range(len(self.clumps)):
            tempClump = self.clumps[i]
            M+= tempClump.mass
            mx += tempClump.mass*tempClump.x
            my += tempClump.mass*tempClump.y
            mz += tempClump.mass*tempClump.z
        return mx/M, my/M, mz/M
    
    def masses(self):
        masses = []
        for i in range(len(self.clumps)):
            masses.append(self.clumps[i].mass)
        return masses
    
    
    def Clumps(self):
        return self.clumps
    
    def Clump(self,i):
        return self.clumps[i]


def init():
    phi = np.linspace(0,2*np.pi,10)
    theta = np.linspace(0,np.pi, 10)
    
    X, Y, Z, Mass, Radius = newCloud.status()
    CMX, CMY, CMZ = newCloud.cm()
    ax.set_xlim3d(CMX-space,CMX+space)
    ax.set_ylim3d(CMY-space,CMY+space)
    ax.set_zlim3d(CMZ-space,CMZ+space)
    
    for i in range(len(X)):
        xm = Radius[i] * np.outer(np.cos(phi), np.sin(theta)) + X[i]
        ym = Radius[i] * np.outer(np.sin(phi), np.sin(theta)) + Y[i]
        zm = Radius[i] * np.outer(np.ones(np.size(phi)), np.cos(theta)) + Z[i]
        ax.plot_surface(xm, ym, zm, rstride=1, cstride=1, color=newCloud.clumps[i].color)

    #ax.draw()
    #ax.show()
    

def animate(i):
    ax.clear()
    Cloud = simulation[i]
    X, Y, Z, Mass, Radius = Cloud.status()
    CMX, CMY, CMZ = Cloud.cm()
    ax.set_xlim3d(CMX-space,CMX+space)
    ax.set_ylim3d(CMY-space,CMY+space)
    ax.set_zlim3d(CMZ-space,CMZ+space)
    phi = np.linspace(0,2*np.pi,10)
    theta = np.linspace(0,np.pi, 10)
    
    #newCloud = newCloud.copy()
    #newCloud.timestep(dt)
    #X, Y, Z, Mass, Radius = newCloud.status()

    for j in range(len(X)):
        xm = Radius[j] * np.outer(np.cos(phi), np.sin(theta)) + X[j]
        ym = Radius[j] * np.outer(np.sin(phi), np.sin(theta)) + Y[j]
        zm = Radius[j] * np.outer(np.ones(np.size(phi)), np.cos(theta)) + Z[j]
        ax.plot_surface(xm, ym, zm, rstride=1, cstride=1, color=Cloud.clumps[j].color)

    #ax.draw()
    #ax.show()
    ax.view_init(elev=20., azim=i)

    
# Set up formatting for the movie files
startTime = time.time()
Writer = animation.writers['ffmpeg']
writer = Writer(fps=20, metadata=dict(artist='Me'), bitrate=18000)
#mywriter = animation.FFMpegWriter()
space = 2000
mass = 10000
n = 200
duration = 5000
dt = 10000
v=0.001
simulation=[]
newCloud = Cloud(N=n, V=v, Space = space, Mass = mass)
simulation.append(newCloud)
for i in range(duration):
    tempCloud=simulation[i].copy()
    tempCloud.timeStep(dt)
    
    simulation.append(tempCloud)


fig = plt.figure()
ax = fig.add_subplot(111,projection='3d')


ani=animation.FuncAnimation(fig, animate, duration, interval=1, init_func = init, blit=False)
Timestamp = ('{:%Y-%b-%d_%H:%M:%S}'.format(datetime.datetime.now()))
ani.save("GravEx%iParticles_%iSpace_%s.mp4"%(n,space, Timestamp), writer=writer)
finishTime = time.time()
#ani.save("GravitationExperiment.mp4")
#plt.show()

print("Total Calculation Time is %0.4f seconds."%(finishTime-startTime))
```
